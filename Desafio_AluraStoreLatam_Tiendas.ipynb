### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
df = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
df.head()
```

```python
df.info()
```

#1. Análisis de facturación



```python
tienda["tienda"] = "Tienda 1"
tienda2["tienda"] = "Tienda 2"
tienda3["tienda"] = "Tienda 3"
tienda4["tienda"] = "Tienda 4"
```

```python
df = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
print("Total de filas:", df.shape[0])
df.head()
```

```python
df["facturacion"] = df["Precio"]
df['facturacion'] = df['facturacion'].round(2)
df.head()
```

```python
facturacion_por_tienda = (
    df.groupby("tienda", as_index=False)["facturacion"]
      .sum()
      .sort_values(by="facturacion", ascending=False)
)

facturacion_por_tienda
```

```python
pd.options.display.float_format = '{:,.2f}'.format
facturacion_por_tienda
```

# 2. Ventas por categoría

```python
categoria_col = 'Categoría del Producto'
tienda_col = 'tienda'
```

```python
ventas_categoria_tienda = (
    df.groupby([tienda_col, categoria_col])
      .size()
      .reset_index(name='cantidad_ventas')
      .sort_values([tienda_col, 'cantidad_ventas'], ascending=[True, False])
)

print("Tabla: cantidad de ventas por categoría y tienda\n")
display(ventas_categoria_tienda)

print("\nTop 5 categorías por cada tienda:")
for t in ventas_categoria_tienda[tienda_col].unique():
    top5 = ventas_categoria_tienda[ventas_categoria_tienda[tienda_col] == t]\
           .sort_values('cantidad_ventas', ascending=False).head(5)
    print(f"\n{t}:")
    display(top5)
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio_tienda = df.groupby('tienda')['Calificación'].mean().reset_index()
```

```python
calificacion_promedio_tienda = calificacion_promedio_tienda.sort_values(by='Calificación', ascending=False)

calificacion_promedio_tienda
```

# 4. Productos más y menos vendidos

```python
ventas_producto_tienda = (
    df.groupby(['tienda', 'Producto'])
      .size()
      .reset_index(name='cantidad_ventas')
)
```

```python
mas_vendidos = ventas_producto_tienda.loc[
    ventas_producto_tienda.groupby('tienda')['cantidad_ventas'].idxmax()
]
```

```python
menos_vendidos = ventas_producto_tienda.loc[
    ventas_producto_tienda.groupby('tienda')['cantidad_ventas'].idxmin()
]
```

# 5. Envío promedio por tienda

```python
envio_promedio_tienda = df.groupby('tienda')['Costo de envío'].mean().reset_index()
```

```python
envio_promedio_tienda = envio_promedio_tienda.sort_values(by='Costo de envío', ascending=False)

print(envio_promedio_tienda)
```

# 6. Graficos

## *Grafico de Facturación por Tienda*

```python
import matplotlib.pyplot as plt

facturacion_por_tienda = facturacion_por_tienda.sort_values(by='facturacion', ascending=True)

plt.figure(figsize=(8, 5))
bars = plt.barh(facturacion_por_tienda['tienda'], facturacion_por_tienda['facturacion'], color='skyblue')
plt.xlabel('Facturación')
plt.title('Facturación por Tienda')

for bar in bars:
    width = bar.get_width()
    plt.text(width - (width * 0.05), bar.get_y() + bar.get_height() / 2,
             f'{width:,.2f}', ha='right', va='center', color='white', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
plt.savefig('Facturación_por_Tienda.png', dpi=300, bbox_inches='tight')
```

# *Ventas por Categoría*

```python
plt.figure(figsize=(12, 6))

categorias = ventas_categoria_tienda[categoria_col].unique()
tiendas = ventas_categoria_tienda[tienda_col].unique()

bar_width = 0.2
x = range(len(categorias))

for i, tienda in enumerate(tiendas):
    datos_tienda = ventas_categoria_tienda[ventas_categoria_tienda[tienda_col] == tienda]
    plt.bar([p + bar_width*i for p in x],
            datos_tienda['cantidad_ventas'],
            width=bar_width,
            label=tienda)

plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Ventas')
plt.title('Ventas por Categoría en Cada Tienda')
plt.xticks([p + bar_width*(len(tiendas)/2 - 0.5) for p in x], categorias, rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.savefig('Ventas_por_Categoría_en_Cada_Tienda.png', dpi=300, bbox_inches='tight')
```

# *Calificación promedio de la tienda*

```python
calificacion_promedio_tienda = df.groupby('tienda')['Calificación'].mean().reset_index()

calificacion_promedio_tienda['Porcentaje'] = (calificacion_promedio_tienda['Calificación'] /
                                              calificacion_promedio_tienda['Calificación'].sum()) * 100

labels = calificacion_promedio_tienda['tienda']
sizes = calificacion_promedio_tienda['Porcentaje']
colors = plt.cm.Paired(range(len(labels)))

fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops=dict(width=0.3)
)

ax.set(aspect="equal")
plt.title("Porcentaje de Calificación Promedio por Tienda")
plt.show()
```

```python
plt.savefig('Porcentaje_de_Calificación_Promedio_por_Tienda.png', dpi=300, bbox_inches='tight')
```

# *Productos más y menos vendidos*

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].barh(mas_vendidos['tienda'], mas_vendidos['cantidad_ventas'], color='green')
axes[0].set_title('Productos más vendidos por tienda')
for i, (prod, cant) in enumerate(zip(mas_vendidos['Producto'], mas_vendidos['cantidad_ventas'])):
    axes[0].text(cant, i, f"{prod} ({cant})", va='center', fontsize=9)

axes[1].barh(menos_vendidos['tienda'], menos_vendidos['cantidad_ventas'], color='red')
axes[1].set_title('Productos menos vendidos por tienda')
for i, (prod, cant) in enumerate(zip(menos_vendidos['Producto'], menos_vendidos['cantidad_ventas'])):
    axes[1].text(cant, i, f"{prod} ({cant})", va='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
plt.savefig('Productos_mas_y_menos_vendidos_por_Tienda.png', dpi=300, bbox_inches='tight')
```

# *Envío promedio por tienda*

```python
plt.figure(figsize=(8, 5))
bars = plt.barh(envio_promedio_tienda['tienda'], envio_promedio_tienda['Costo de envío'], color='orange')
plt.xlabel('Costo promedio de envío')
plt.title('Costo Promedio de Envío por Tienda')

for bar in bars:
    width = bar.get_width()
    plt.text(width - (width * 0.1), bar.get_y() + bar.get_height() / 2,
             f'{width:.2f}', ha='right', va='center', color='white', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
plt.savefig('Costo_Promedio_de_Envío_por_Tienda.png', dpi=300, bbox_inches='tight')
```
